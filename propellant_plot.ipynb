```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df_sorted = pd.read_csv("combustion_sorted.csv")
df = pd.read_csv("combustion.csv")
```

```python
print(df)
```

```python
ax = sns.lineplot(data=df, x="Pc", y="Isp")
```

```python
df_ethanol = df[df["fuel"] == "C2H5OH"]
df_methane = df[df["fuel"] == "methane"]
df_propane = df[df["fuel"] == "propane"]
```

```python
print(df_ethanol)
```

```python
sns.set(rc={"figure.dpi":300, 'savefig.dpi':300})
sns.set_context('notebook')
sns.set_style("ticks")
ax1 = sns.lineplot(data=df_ethanol, x="Pc", y="Isp",hue="fuel",palette="Dark2_r")
ax2 = sns.lineplot(data=df_methane, x="Pc", y="Isp",hue="fuel", palette="Greens_r")
ax3 = sns.lineplot(data=df_propane, x="Pc", y="Isp",hue="fuel", palette="RdYlBu")
```
